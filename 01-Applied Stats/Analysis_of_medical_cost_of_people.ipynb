### 4.1 Let's check if the charges of people who smoke differ significantly from the people who don't

We will first establish the Null and Alternate Hypothesis as follows : <br/><br/>
<b>Ho = charges of smokers and non-smokers do not differ</b> <br>
<b>Ha = charges of smokers and non-smokers vary significantly</b> <br>
<b>Significance level = 5%</b> 

```python
print (df['smoker'].value_counts())
```

```python
nonsmoker_grp = df[df['smoker'] == 'no']
smoker_grp    = df[df['smoker'] == 'yes']
```

```python
from scipy.stats import ttest_ind

# two-sample t-test
t_statistic, p_value = ttest_ind(nonsmoker_grp['charges'], smoker_grp['charges'])
print ('t_statistic value : ', t_statistic)
print ('p_value           : ', p_value, '\n\n')

# p_value < 0.05 => alternative hypothesis:
# they don't have the same mean at the 5% significance level
if p_value < 0.05 :
    print (f'We reject the Null Hypothesis since {p_value} < 0.05 ')
else :
    print (f'We fail to reject the null Hypothesis since {p_value} >= 0.05')
```

<b> Observations :</b> <br/>
1. The charges paid by smokers is <b>46 standard deviations more</b> than that paid by non smokers <br/> 
2. Since p value is much less than 0.05 we <b>reject the null hypothesis at 95% confidence level</b> <br/>
3. This implies we <b>accept the alternate hypothesis</b> that there is a <b>significant difference in the charges paid by smokers compared to non-smokers</b> <br/>
4. The same is visualized in the scatter plot below 

```python
plt.figure(figsize=(10,8))
sns.scatterplot(x=df['age'], y=df['charges'], hue=df['smoker']) ;
```

### 4.2 Let's check if the BMI of males differ significantly from that of females

Again, we establish the Null and Alternate Hypothesis as follows : <br/><br/>
<b>Ho = Gender has no impact on BMI</b> <br>
<b>Ha = Gender has an impact on BMI</b> <br>
<b>Significance level = 5%</b> 


```python
print (df['sex'].value_counts())
```

```python
males_grp   = df[df['sex'] == 'male']
females_grp = df[df['sex'] == 'female']
```

```python
from scipy.stats import ttest_ind

# two-sample t-test
t_statistic, p_value = ttest_ind(males_grp['bmi'], females_grp['bmi'])
print ('t_statistic value : ', t_statistic)
print ('p_value           : ', p_value, '\n\n')

# p_value < 0.05 => alternative hypothesis:
# they don't have the same mean at the 5% significance level
if p_value < 0.05 :
    print (f'We reject the Null Hypothesis since {p_value} < 0.05 ')
else :
    print (f'We fail to reject the null Hypothesis since {p_value} >= 0.05')
```

<b> Observations : </b> <br/>
1. The BMI of males is <b>1.5 standard deviations more</b> than that of females <br/>
2. Since p value is greater than 0.05 we <b>fail to reject the null hypothesis at 95% confidence level</b> <br/>
3. This implies <b>Gender has no major imapct on the BMI</b> <br/>
4. The same is visualized in the scatter plot below 

```python
plt.figure(figsize=(10,8))
sns.scatterplot(x=df['age'], y=df['bmi'], hue=df['sex']) ;
```

### 4.3 Let's check if the proportion of smokers is significantly different in different genders

Again, we establish the Null and Alternate Hypothesis as follows : <br/><br/>

<b>Ho = Proportion of smokers in males and females is same</b> <br>
<b>Ha = Proportion of smokers in males and females is different</b> <br> 

```python
female_smokers = df[df['sex'] == 'female'].smoker.value_counts()[1]  # number of female smokers
male_smokers   = df[df['sex'] == 'male'].smoker.value_counts()[1]    # number of male smokers
n_females = df.sex.value_counts()[1]                                 # number of females in the data
n_males   = df.sex.value_counts()[0]                                 #number of males in the data

print ('Total number of females           : ', n_females)
print ('Total number of females who smoke : ', female_smokers)
print ('\n')
print ('Total number of males             : ', n_males)
print ('Total number of males who smoke   : ', male_smokers)
```

```python
print(f' Proportion of smokers in females, males = {round(115/662,2)}%, {round(159/676,2)}% respectively')
```

The proportions are different but are they statistically significant?

```python
from statsmodels.stats.proportion import proportions_ztest

zstat, pval = proportions_ztest([female_smokers, male_smokers] , [n_females, n_males])

print ('z_statistic value : ', zstat)
print ('p_value           : ', pval, '\n\n')

if pval < 0.05:
    print(f'With a p-value of {round(pval,4)} the difference is significant. aka |We reject the null|')
else:
    print(f'With a p-value of {round(pval,4)} the difference is not significant. aka |We fail to reject the null|')
```

<b> Observations : </b><br/>
1. The proportion of smokers in males is <b>2.7 standard deviations more</b> than that of females  <br/>
2. Since p value is less than 0.05 we <b>reject the null hypothesis at 95% confidence level</b> <br/>
3. This implies <b>Proportion of smokers in males is significantly different from that of the females</b> 

### 4.4 Let's check if the distribution of bmi across women with no children, one child and two children, the same

 We establish the Null and Alternate Hypothesis as follows : <br/><br/>
<b>Ho = No of children has no effect on BMI of females</b> <br>
<b>Ha = No of children has an effect on BMI of females</b> <br> 

```python
from scipy.stats import f_oneway

female_grp = df[df['sex'] == 'female']

no_child_bmi  = female_grp[female_grp['children'] == 0]['bmi']
one_child_bmi = female_grp[female_grp['children'] == 1]['bmi']
two_child_bmi = female_grp[female_grp['children'] == 2]['bmi']

f_stat, p_value = f_oneway(no_child_bmi, one_child_bmi, two_child_bmi)
print ('p_value           : ', p_value, '\n\n')

if p_value < 0.05:
    print(f'With a p-value of {round(p_value,4)} the BMI variation is significant. aka |We reject the null|')
else:
    print(f'With a p-value of {round(p_value,4)} the BMI variation is not significant. aka |We fail to reject the null|')
```

<b> Observations : </b><br/>
1. Since p value is much greater than 0.05 we <b>fail to reject the null hypothesis at 95% confidence level</b><br/>
3. This implies <b>Number of children has no effect on BMI of females</b>
